```python
#作用域：L：局部作用域，函数定义的变量；E：嵌套父级函数的局部作用域；G：全局变量，模块级别定义的变量；B：系统固定模块里面的变量
x = 2         # 全局的
def f():
    x = 3     # 局部的
    return x
print(x)      # 结果为2
print(f())    # 结果为3
```

```python
# 先对局部作用域进行搜索，然后对全局作用域进行搜索，如果其中两个作用域不能找到，最后会在作用域搜索内建。
x = 5
def f():
    y = 2*x       
    return y
print(f())         # 使用的是global scope x
```

```python
#嵌套函数：函数中的函数
def square():
    def add():
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square()) 
```

#默认参数、可变参数、关键字参数
def f(a,b=1): #已经初始给定的参数
def f(*args): #值可以是任意多个
def f(** kwargs) #kwargs是个字典

```python
# 默认参数， 必选参数在前，默认参数在后，否则python解释器会报错
# 默认参数一定要指向不变对象 
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# 当我们想改变默认参数时，直接把值放进去
print(f(5,4,3))
```

```python
# 可变参数 *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,5)
```

```python
#Lambda函数也叫匿名函数，允许快速定义单行函数。通常是在需要一个函数，但是又不想费神去命名一个函数的场合下使用，也就是指匿名函数。
# lambda函数
# 单个参数的
square = lambda x: x**2    
print(square(4))

# 多个参数的
tot = lambda x,y,z: x+y+z   
print(tot(1,2,3))
```

```python
number_list = [1,2,3]
# map(func,seq):对列表中的所有项应用函数
y = map(lambda x:x**2,number_list)
print(list(y))
```

```python
#迭代器
name = "ncxhdc"
it = iter(name)
print(next(it))    # 打印容器中的下一个值
print(*it)         # 打印容器中的剩余值
```

```python
#zip()函数用于将可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的对象，这样做的好处是节约了不少内存。
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2)  # 返回一个对象
print(z)
z_list = list(z)      # list()转换为列表
print(z_list)
```

```python
# 与 zip 相反，zip(*) 可理解为解压，返回二维矩阵
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip)     # un_zip 返回一个 tuple
print(un_list1)
print(un_list2)
print(type(un_list2))
```

```python
#（列表推导式）列表解析式是一种非常简洁的创建列表的方式。
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)
```

```python
# 使用if语句进行过滤
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
